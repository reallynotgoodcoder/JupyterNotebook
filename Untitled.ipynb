```python
# Восьме завдання:
def average_gdp_last_10_years(df):
    gdp_columns = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    
    df['Average GDP'] = df[gdp_columns].mean(axis=1)
    
    return df['Average GDP'].sort_values(ascending=False)

print(average_gdp_last_10_years(merged_data))
```

```python
# Дев'яте завдання:
def gdp_change_for_5th_country(df):
    gdp_columns = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    
    df['Average GDP'] = df[gdp_columns].mean(axis=1)
    
    fifth_country = df['Average GDP'].nlargest(5).idxmin()
    
    initial_gdp = df.loc[fifth_country, '2006']
    final_gdp = df.loc[fifth_country, '2015']
    gdp_change = final_gdp - initial_gdp
    
    return (fifth_country, gdp_change)

gdp_change = gdp_change_for_5th_country(merged_data)
print(gdp_change)
```

```python
# Десяте завдання
def max_renewable_country(df):
    max_country = df['% Renewable'].idxmax()
    max_value = df.loc[max_country, '% Renewable']
    
    return (max_country, max_value)

max_renewable = max_renewable_country(merged_data)
print(max_renewable)
```

```python
# Одинадцяте завдання
def sixth_largest_population(df):
    if 'Energy Supply' not in df.columns or 'Energy Supply per Capita' not in df.columns:
        raise ValueError("Відсутні потрібні стовпці у датафреймі!")

    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
    df = df.dropna(subset=['Population'])
    sorted_df = df.sort_values(by='Population', ascending=False)

    if len(sorted_df) < 6:
        print("У датафреймі немає шостої країни за населенням!")
        if len(sorted_df) >= 5:
            print("Повертаємо п’яту країну:")
            sixth_country = sorted_df.iloc[4].name
            sixth_population = sorted_df.iloc[4]['Population']
        else:
            raise IndexError("У датафреймі занадто мало країн для аналізу!")
    else:
        sixth_country = sorted_df.iloc[5].name
        sixth_population = sorted_df.iloc[5]['Population']

    return (sixth_country, sixth_population)

sixth_pop = sixth_largest_population(merged_data)
print(sixth_pop)
```

```python
# Дванадцяте завданная
def citations_energy_correlation(df):
    required_columns = ['Citable documents', 'Population', 'Energy Supply per Capita']
    if not all(col in df.columns for col in required_columns):
        raise ValueError("Відсутні необхідні стовпці у датафреймі!")

    df['Citable docs per capita'] = df['Citable documents'] / df['Population']
    correlation = df[['Citable docs per capita', 'Energy Supply per Capita']].corr().iloc[0, 1]

    return correlation

correlation_result = citations_energy_correlation(merged_data)
print(correlation_result)
```

```python
# Тринадцяте завдання
def renewable_energy_flag(df):
    if '% Renewable' not in df.columns:
        raise ValueError("Відсутній стовпець '% Renewable' у датафреймі!")

    median_value = df['% Renewable'].median()
    df['Renewable Flag'] = (df['% Renewable'] >= median_value).astype(int)

    return df['Renewable Flag'].sort_values(ascending=True)

renewable_flag_series = renewable_energy_flag(merged_data)
print(renewable_flag_series)
```

```python
# Чотирнадцяте завдання
def continent_population_stats(df):
    ContinentDict = {
        'China': 'Asia',
        'United States': 'North America',
        'Japan': 'Asia',
        'United Kingdom': 'Europe',
        'Russian Federation': 'Europe',
        'Canada': 'North America',
        'Germany': 'Europe',
        'India': 'Asia',
        'France': 'Europe',
        'South Korea': 'Asia',
        'Italy': 'Europe',
        'Spain': 'Europe',
        'Iran': 'Asia',
        'Australia': 'Australia',
        'Brazil': 'South America'
    }

    if 'Country' in df.columns:
        df.set_index('Country', inplace=True)

    df['Continent'] = df.index.map(ContinentDict)
    df = df.dropna(subset=['Continent'])
    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
    result = df.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])

    return result

continent_population_df = continent_population_stats(merged_data)
continent_population_df
```
