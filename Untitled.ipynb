# Інформація про Джибуті

Джибуті — це маленька країна на сході Африки, розташована на узбережжі Червоного моря та Баб-ель-Мандебської протоки.

## Площа
Площа Джибуті складає приблизно **23 200 км²**.  

## Населення
<p style="color:blue;">Населення Джибуті становить приблизно <strong>900 000 осіб</strong> (на 2023 рік).</p>

## Державний устрій
Джибуті є _**президентською республікою**_. Президент є головою держави та уряду.

## Гіперпосилання

- Вікіпедія про Джибуті: [Wikipedia](https://en.wikipedia.org/wiki/Djibouti)  


```python
def BubbleSort(numbers):
    numberOfPermutations = 0

    for i in range(len(numbers)):
        for j in range(len(numbers) - i - 1):
            if numbers[j] > numbers[j + 1]:
                temp = numbers[j]
                numbers[j] = numbers[j + 1]
                numbers[j + 1] = temp
                
                numberOfPermutations += 1

    return numberOfPermutations

def CheckValue(numbers):
    for i in range(len(numbers)):
        if numbers[i].isnumeric() == False:
            return False

    return True

numbers = list(input("Введіть числа через пробіл: ").split())

if CheckValue(numbers) == False:
    print("Помилка. У списку присутнє нечислове значення")
else:
    numbers = list(map(int, numbers))
    print("Введені числа: ", numbers, "\n")
    
    numberOfPermutations = BubbleSort(numbers)
    
    print("Відсортовані числа: ", numbers)
    print("Кількість перестановок: ", numberOfPermutations)
```
